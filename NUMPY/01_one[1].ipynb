```python
import numpy as np
data = np.loadtxt('populations.txt')
```

```python
data
```

```python
year,hares,lynxes,carrots = data.T         #columns to variables , .T means transpose of a matrix
print(year)
```

```python
# the mean population over time, 
# 1st shows as row and 2nd shows as column
population = data[:,1:]
population
```

```python
# standard deviation row-wise    = axis-1
# standard deviation column-wise    = axis-0
population.std(axis=0)
```

```python
population.std(axis=1)
```

```python
# which species has the highest population each year 
np.argmax(population,axis=1)          #row-wise
```
